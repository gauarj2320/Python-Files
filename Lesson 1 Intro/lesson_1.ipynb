- Object Oriented Language: Everything in python is an object

## <span style="color:yellow">Disadvantages of Python</span>

## <span style="color:#33ff00">Identifiers in Python</span>
- Identifiers: Name given to python objects like  variable,class,object,function,package etc.

- Rules to define Identifiers in Python:
    - Name can be any combination of a-z,A-z,0-1 and '_'
    - Name in Python are case sensitive
    - Names cannot begin with a number
    - Python Keywords cannot be used as names

- Python Keywords List:
